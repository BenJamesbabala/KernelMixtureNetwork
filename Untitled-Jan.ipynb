```python
def determine_centers(train_y, method='all', d=1, k=100, include_borders = True):
    
    ## Keep all points as kernel centers
    if method == 'all':
        return train_y.ravel()
    
    ## If include_borders always use outer points to keep expressiveness at the borders
    train_y = sorted(train_y)
    if include_borders:
        centers = [train_y[0], train_y[-1]]
        train_y = train_y[1:-1]
    else:
        centers = []
    
    ## Iteratively remove part of pairs that are closest together until everything is at least 'd' apart 
    if method == 'distance':
        cluster_centers = []
        if include_borders:
            last_point = centers[0]
        else:
            last_point = -float("inf")
        for y in train_y:
            if y-last_point >= d:
                last_point = y
                cluster_centers.append(y)
        cluster_centers = np.array(cluster_centers)
    
    ## Use 1-D k-means clustering to determine k output points plus the two end points
    if method == 'k_means':
        kmeans = KMeans(n_clusters=k, n_jobs=-2)
        kmeans.fit(np.array(train_y).reshape(-1, 1))
        cluster_centers = kmeans.cluster_centers_
        
    ## Use agglomerative clustering to determine k output points plus the two end points
    #if method is 'agglomerative':
    #    model = AgglomerativeClustering(n_clusters=k)
    #    model.fit(train_y)
    #    cluster_centers = model.cluster_centers_
        
    ## Concatenate the cluster centers and the borders and return them
    return np.append(cluster_centers.ravel(), np.array(centers).ravel())
```

```python
def build_toy_dataset(nsample=40000):
    y_data = np.float32(np.random.uniform(-10.5, 10.5, (1, nsample))).T
    r_data = np.float32(np.random.normal(size=(nsample,1))) # random noise
    x_data = np.float32(np.sin(0.75*y_data)*7.0+y_data*0.5+r_data*1.0)
    return train_test_split(x_data, y_data, random_state=42, train_size=0.5)

X_train, X_test, y_train, y_test = build_toy_dataset()

sns.regplot(X_train, y_train, fit_reg=False)
```

```python
class KernelMixtureNetwork(object):
    def __init__(self, sigmas, method='all', d=1, k=100, include_borders=True, batch_size=32, n_epoch=10):
        ## Sigmas for the Gaussian kernel
        self.sigmas = sigmas
        ## Method to downsample kernel centers
        self.method = method
        ## Parameters for different methods, d = distance, k = number clusters, include_borders = to always keep extremes
        self.d = d
        self.k = k
        self.include_borders = include_borders
        
        ## Training parameters
        self.batch_size = batch_size
        self.n_epoch = n_epoch
    
    def fit(self, X, y, eval_set=None):
        if eval_set is None:
            X_test, y_test = X, y
        else:
            X_test, y_test = eval_set
            
        ## Build centers according to method
        self.centers = determine_centers(y, self.method, self.d, self.k, self.include_borders)
        self.n_centers = len(self.centers)
        self._build_network(X.shape[1])
        
        ## Keep track of train and test losses
        self.train_losses, self.test_losses = [], []
        
        indices = np.arange(len(y))
        print('Start fitting')
        for i in range(self.n_epoch):
            
            ## Shuffle indices for batches
            np.random.shuffle(indices)
            
            for b in range(len(y) // self.batch_size):
                cur_indices = indices[b * self.batch_size : (b+1) * self.batch_size]
                cur_X = X[cur_indices]
                cur_y = y[cur_indices]
                feed_dict = {self.X: cur_X, self.y: cur_y}
                ## Fit current batch
                self.sess.run(self.train_op, feed_dict=feed_dict)
                
            # Full train and test loss
            train_loss = self.sess.run(self.mean_loss, {self.X: X, self.y: y})
            test_loss = self.sess.run(self.mean_loss, {self.X: X_test, self.y: y_test})
            self.train_losses.append(train_loss)
            self.test_losses.append(test_loss)
            print("train loss: {:.4f}, test loss: {:.4f}".format(train_loss, test_loss))
            
    def _gaussian_kernel(self, y, y_acc, sigma):
        ## Tensorflow node that calculates kernel distance between a center and current y
        return 1. / (tf.sqrt(2 * pi) * sigma) * tf.exp(-tf.square(y-y_acc) / (2 * tf.square(sigma)))

    def _density_output_function(self, w, y, kernel_centers, sigmas):
        self.output_kernels = []
        for center in kernel_centers:
            for sigma in sigmas:
                self.output_kernels.append(tf.reshape(self._gaussian_kernel(y, center, sigma), [-1, 1]))
        self.output_kernels = tf.concat(self.output_kernels, axis=1)
        print(self.output_kernels.get_shape())
        self.output_nodes = tf.multiply(w, self.output_kernels) #, transpose_b=True
        return tf.reduce_sum(self.output_nodes, axis=1, keep_dims=True) / self.total_weight
    
    def _build_network(self, n_features):
        self.sess = tf.Session()
        self.X = tf.placeholder(tf.float32, [None, n_features])
        self.y = tf.placeholder(tf.float32, [None])
        
        self.W_1 = tf.Variable(tf.random_normal(shape=[n_features, 48]))
        self.b_1 = tf.Variable(tf.random_normal(shape=[48]))
        self.layer_1 = tf.nn.tanh(tf.matmul(self.X, self.W_1) + self.b_1)
        
        self.W_2 = tf.Variable(tf.random_normal(shape=[48, 24]))
        self.b_2 = tf.Variable(tf.random_normal(shape=[24]))
        self.layer_2 = tf.nn.tanh(tf.matmul(self.layer_1, self.W_2) + self.b_2)
        
        self.W_3 = tf.Variable(tf.random_normal(shape=[24, len(self.sigmas) * self.n_centers]))
        self.b_3 = tf.Variable(tf.random_normal(shape=[len(self.sigmas) * self.n_centers]))
        self.w = tf.matmul(self.layer_2, self.W_3) + self.b_3
        self.w = tf.nn.softplus(self.w)
        
        self.total_weight = tf.reduce_sum(self.w, axis=1, keep_dims=True)
        
        print('total_weight', self.total_weight.get_shape())
        
        self.probabilities = self.w / self.total_weight
        
        print('probabilities', self.probabilities.get_shape())
        
        #self.sigmas = tf.nn.softplus(tf.Variable(self.sigmas))

        self.output_layer = self._density_output_function(self.w, self.y, self.centers, self.sigmas)
        
        print('output_layer', self.output_layer.get_shape())
        
        self.loss = -tf.log(self.output_layer)
        self.mean_loss = tf.reduce_mean(self.loss)
        
        self.optimizer = tf.train.AdamOptimizer()
        self.train_op = self.optimizer.minimize(self.mean_loss)
        print('Variable initialization')
        self.sess.run(tf.global_variables_initializer())
        
    def sample(self, X):
        #w = self.sess.run(self.w, feed_dict={self.X: X})
        probs = self.sess.run(self.probabilities, feed_dict={self.X: X})
        samples = []
        for i in range(X.shape[0]):
            kernel_index = np.random.choice(range(len(probs[i])), p=probs[i])
            center_index = kernel_index // len(self.sigmas)
            sigma_index = kernel_index % len(self.sigmas)
            samples.append(np.random.normal(loc=self.centers[center_index], scale=self.sigmas[sigma_index]))
        return samples
        
#kmn = KernelMixtureNetwork([0.2, 0.4, 0.7, 1.], method='k_means', k=10, n_epoch=1000)
#kmn.fit(X_train, y_train.ravel(), (X_test, y_test.ravel()))
```

```python
class KernelMixtureNetwork(object):
    def __init__(self, sigmas, method='all', d=1, k=100, batch_size=32, n_epoch=10):
        self.init_sigmas = [math.log(math.exp(s) - 1) for s in sigmas]
        self.method = method
        self.d = d
        self.k = k
        self.batch_size = batch_size
        self.n_epoch = n_epoch
    
    def fit(self, X, y, eval_set=None):
        if eval_set is None:
            X_test, y_test = X, y
        else:
            X_test, y_test = eval_set
            
        self.centers = determine_centers(y, self.method, self.d, self.k)
        self.n_centers = len(self.centers)
        self._build_network(X.shape[1])
        
        indices = np.arange(len(y))
        print('Start fitting')
        for i in range(self.n_epoch):
            np.random.shuffle(indices)
            for b in range(len(y) // self.batch_size):
                cur_indices = indices[b * self.batch_size : (b+1) * self.batch_size]
                cur_X = X[cur_indices]
                cur_y = y[cur_indices]
                feed_dict = {self.X: cur_X, self.y: cur_y}
                self.sess.run(self.train_op, feed_dict=feed_dict)
            train_loss = self.sess.run(self.mean_loss, {self.X: X, self.y: y})
            test_loss = self.sess.run(self.mean_loss, {self.X: X_test, self.y: y_test})
            cur_sigmas = self.sess.run(self.sigmas, None)
            print("train loss: {:.4f}, test loss: {:.4f}, sigmas: {}".format(train_loss, test_loss, cur_sigmas))
            
    def _gaussian_kernel(self, y, y_acc, sigma):
        return 1. / (tf.sqrt(2 * pi) * tf.transpose(sigma)) * tf.exp(-tf.square(y-y_acc) / (2 * tf.transpose(tf.square(sigma))))

    def _density_output_function(self, w, y, kernel_centers, sigmas):
        self.output_kernels = []
        for center in kernel_centers:
            #self.output_kernels.append(tf.reshape(self._gaussian_kernel(y, center, sigmas), [-1, 1]))
            self.output_kernels.append(self._gaussian_kernel(y, center, sigmas))
            #print(self.output_kernels[-1].get_shape())
        self.output_kernels = tf.transpose(tf.concat(self.output_kernels, axis=0))
        #print('output_kernels', self.output_kernels.get_shape())
        self.output_nodes = tf.multiply(w, self.output_kernels) #, transpose_b=True
        #print('output_nodes', self.output_nodes.get_shape())
        return tf.reduce_sum(self.output_nodes, axis=1, keep_dims=True) / self.total_weight
    
    def _build_network(self, n_features):
        self.sess = tf.Session()
        self.X = tf.placeholder(tf.float32, [None, n_features])
        self.y = tf.placeholder(tf.float32, [None])
        
        self.W_1 = tf.Variable(tf.random_normal(shape=[n_features, 48]))
        self.b_1 = tf.Variable(tf.random_normal(shape=[48]))
        self.layer_1 = tf.nn.tanh(tf.matmul(self.X, self.W_1) + self.b_1)
        #self.bn_layer_1 = tf.contrib.layers.batch_norm(self.layer_1)
        
        self.W_2 = tf.Variable(tf.random_normal(shape=[48, 24]))
        self.b_2 = tf.Variable(tf.random_normal(shape=[24]))
        self.layer_2 = tf.nn.tanh(tf.matmul(self.layer_1, self.W_2) + self.b_2)
        
        self.W_3 = tf.Variable(tf.random_normal(shape=[24, 24]))
        self.b_3 = tf.Variable(tf.random_normal(shape=[24]))
        self.layer_3 = tf.nn.tanh(tf.matmul(self.layer_2, self.W_3) + self.b_3)
        
        #self.W_4 = tf.Variable(tf.random_normal(shape=[24, 12]))
        #self.b_4 = tf.Variable(tf.random_normal(shape=[12]))
        #self.layer_4 = tf.nn.tanh(tf.matmul(self.layer_3, self.W_4) + self.b_4)
        
        self.W_5 = tf.Variable(tf.random_normal(shape=[24, len(self.init_sigmas) * self.n_centers]))
        self.b_5 = tf.Variable(tf.random_normal(shape=[len(self.init_sigmas) * self.n_centers]))
        self.w = tf.matmul(self.layer_3, self.W_5) + self.b_5
        self.w = tf.nn.softplus(self.w)
        
        #print(self.w.get_shape())
        
        self.total_weight = tf.reduce_sum(self.w, axis=1, keep_dims=True)
        self.probabilities = self.w / self.total_weight
        
        self.sigmas = tf.nn.softplus(tf.Variable([self.init_sigmas]))
        
        #print(self.sigmas.get_shape())

        self.output_layer = self._density_output_function(self.w, self.y, self.centers, self.sigmas)
        
        #print(self.output_layer.get_shape())
        
        self.loss = -tf.log(self.output_layer)
        
        #print(self.loss.get_shape())
        
        self.mean_loss = tf.reduce_mean(self.loss)
        
        #self.reg_loss = self.loss  + 0.001 *(tf.nn.l2_loss(self.W_1) + tf.nn.l2_loss(self.W_2) + tf.nn.l2_loss(self.W_3) + \
        #             tf.nn.l2_loss(self.W_4) + tf.nn.l2_loss(self.W_5))
        self.reg_loss = self.loss  + 0.001 *(tf.nn.l2_loss(self.W_1) + tf.nn.l2_loss(self.W_2) + tf.nn.l2_loss(self.W_5))
        
        self.mean_reg_loss = tf.reduce_mean(self.reg_loss)
        
        self.optimizer = tf.train.AdamOptimizer()
        self.train_op = self.optimizer.minimize(self.mean_reg_loss)
        print('Variable initialization')
        self.sess.run(tf.global_variables_initializer())
        
    def sample(self, X):
        #w = self.sess.run(self.w, feed_dict={self.X: X})
        probs = self.sess.run(self.probabilities, feed_dict={self.X: X})
        samples = []
        sigmas = self.sess.run(self.sigmas).ravel()
        for i in range(X.shape[0]):
            kernel_index = np.random.choice(range(len(probs[i])), p=probs[i])
            center_index = kernel_index // len(self.init_sigmas)
            sigma_index = kernel_index % len(self.init_sigmas)
            samples.append(np.random.normal(loc=self.centers[center_index], scale=sigmas[sigma_index]))
        return samples
        
kmn = KernelMixtureNetwork([0.2, 0.05, 0.1, 0.5], method='k_means', k=100, n_epoch=400)
```
